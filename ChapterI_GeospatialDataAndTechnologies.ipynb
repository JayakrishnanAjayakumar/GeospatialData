## So what exactly is Geospatial Data?

The word **geospatial is used to indicate that data that has a geographic component to it**. Or simply it means that the **data can be located at the surface of earth or can be mapped!!!**

Let's look at some examples

![Geo1](images/geo1.png)

**So any data that has explicit location information (like coordinates) or location information that can be extracted out (like address), can be considered Geospatial data**.

So what do you think when some one says

**"Hey I am right now in Woods building, CWRU"**........

Is that Geospatial data

For a more formal definition

> **Geospatial data is information that describes objects, events or other features with a location on or near the surface of the earth**. Geospatial data typically **combines location information (usually coordinates on the earth) and attribute information (the characteristics of the object, event or phenomena concerned) with temporal information (the time or life span at which the location and attributes exist)**. The location provided may be static in the short term (for example, the location of a piece of equipment, an earthquake event, children living in poverty) or dynamic (for example, a moving vehicle or pedestrian, the spread of an infectious disease).

## Is Spatial Data and Geospatial Data the same??

The term **spatial comes from the latin word spatium which means 'space'. Spatial, therefore, refers to anything that occupies space or the study of the distribution of things in space**.
So spatial data can refer to data about spatial layout of the human body, or a room or a car or virtually anything. 

![Geo2](images/geo2.png)

The image shown above is a valid example of spatial data that's not in geographical coordinate system. 

Geospatial is the combination of 'Geo' and 'Spatial' and 'Geo' indicates that the focus is on features relating to Earth.

If you were to look for spatial patterns of Covid-19 spread in University Heights a county level aggregated Covid-19 case data for Cuyahoga county is probably not so much useful. Depending on the real-world issue you are trying to solve, you would need geospatial data at different scales. 

In the beginning of the video Dr Curtis mentions about data that are not collected by our team. Such data are called **secondary data**. 

Majority of spatial research work revolves around using secondary data from official sources, hospitals (as explained in the video), web repositories, and spatially enabled social media sources. 

While its relatively easy to get access to such data sources (barring the health data from hospitals), it is hard to understand the **nuances behind the data collection strategy (as it was conducted by someone else)**. Data analysis (both spatial and non-spatial) becomes a hard task when there is a lack of information about how the data was assimilated (what was the sampling strategy used, was there any data gaps). 

In contrast **primary data are collected by our team** through various ways including surveys, spatial video (this is a technology that we are heavily using for collecting fine scale spatial data), gps enabled IoT deivces. Since we have collected the primary data, we will have a better understanding of the nuances associated with the data. **But the cost and effort required to collect primary data can be overwhelming for many research initiatives**. 

We will look into the details of a particular primary data collection strategy called Spatial Video (SV) surveying in the next chapter. 


One another key insight mentioned in the video is about the **power of programming and automations**. This is especially important when you are working in a health settings where quick insights can literally save lives. Any near real-time disease survelliance would require completely automated pipelines, so that the subject matter experts can focus on the task of "making sense of the data" rather than handling data challenges. 

Last but not least, while handling any spatial data, one sohuld be aware of the **privacy and confidentiality risks** associated with the data. We have made great advances in the field of spatial and social sciences that we can now literally combine any spatial data with any other socio-spatial data through the power of spatial-joins. While this is fantastic news for research, it opens up a can of worms interms of privacy and confidentiality.
